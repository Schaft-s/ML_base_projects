```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras import layers
```

```python
df = pd.read_csv('final.csv')
df
```

```python
df = df.applymap(lambda x: pd.to_numeric(x, errors='coerce'))
df = df.dropna()
df
```

```python
df = df.iloc[:, 1:]
```

```python
df
```

```python
# Разделение признаков и таргетов
X = df.iloc[:, :1000]  # Первые 1000 столбцов - признаки
Y = df.iloc[:, 1000:]  # Последние 36 столбцов - таргеты
```

```python
Y = Y[['C11', 'C12','C13', 'C14', 'C15', 'C16','C22','C23', 'C24', 'C25', 'C26','C33','C34',  'C35','C36','C44','C45','C46', 'C55','C56','C66']]
```

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
y_normalized = scaler.fit_transform(Y)
len(y_normalized)
```

```python
# Разбиение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y_normalized, test_size=0.1, random_state=42)

# Теперь у вас есть X_train, X_test, y_train, y_test для дальнейшего использования
```

```python
model = keras.Sequential()
model.add(layers.Input(shape=(X_train.shape[1],)))  # Используем Input как первый слой
model.add(layers.Dense(128, activation='relu'))  # Входной слой
model.add(layers.Dense(64, activation='relu'))  # Скрытый слой
model.add(layers.Dense(Y.shape[1], activation='linear'))  # Выходной слой
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae', 'mse'])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1)
```

```python
loss, mae, mse = model.evaluate(X_test, y_test)

print(f"Test Loss: {loss:.4f}")
print(f"Test MAE: {mae:.4f}")
```

```python
# Создание модели
model = keras.Sequential()
model.add(layers.Input(shape=(X_train.shape[1],)))  # Входной слой
model.add(layers.Dense(128, activation='selu'))  # Первый скрытый слой
model.add(layers.Dropout(0.2))  # Дропаут с вероятностью 0.5
model.add(layers.Dense(64, activation='selu'))  # Второй скрытый слой
model.add(layers.Dropout(0.2))  # Дропаут с вероятностью 0.5
model.add(layers.Dense(Y.shape[1], activation='linear'))  # Выходной слой
```

```python
model.summary()
```

```python
# Компиляция модели
model.compile(optimizer='nadam', loss='mean_squared_error', metrics=['mae', 'mse'])

# Обучение модели
model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.1)
```

```python
# Оценка модели на тестовых данных
loss, mae, mse = model.evaluate(X_test, y_test)

print(f"Test MAE: {mae:.4f}")
print(f"Test MSE: {mse:.4f}")
```
