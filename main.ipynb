```python
df = pd.read_csv('output.csv')
df = df.iloc[:, 1:]
df
```

```python
df2 = pd.read_csv('new.csv')
df2 = df2.iloc[:, 3:]
df2
```

```python
dd = pd.concat([df, df2])
dd
```

```python
dd.drop_duplicates()
dd
```

```python
dd.to_csv('final.csv')
```

```python
# Разделение признаков и таргетов
X = dd.iloc[:, :1000]  # Первые 1000 столбцов - признаки
Y = dd.iloc[:, 1000:]  # Последние 36 столбцов - таргеты
```

```python
Y.iloc[0]
```

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
y_normalized = scaler.fit_transform(Y)
len(y_normalized)
```

```python
# Разбиение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y_normalized, test_size=0.1, random_state=42)

# Теперь у вас есть X_train, X_test, y_train, y_test для дальнейшего использования
```

```python
len(y_test)
```

```python
X_train_1 = X_train.values.reshape(-1, 10, 10, 10, 1)
X_test_1 = X_test.values.reshape(-1, 10, 10, 10, 1)
```

```python
len(X_train_1[0][0][0][0])
```

```python
import tensorflow as tf

# Создание модели
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(10, 10, 10, 1)),
    tf.keras.layers.Conv3D(filters=32, kernel_size=3, activation='selu'),
    tf.keras.layers.MaxPooling3D(pool_size=2),
    tf.keras.layers.Conv3D(filters=64, kernel_size=3, activation='selu'),
    tf.keras.layers.MaxPooling3D(pool_size=2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=128, activation='selu'),
    tf.keras.layers.Dense(units=36)  # Выходной слой для регрессии, 36 чисел
])

# Компиляция модели
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error', 'mean_squared_error' ])

# Вывод структуры модели
model.summary()
```

```python
# Обучение модели
model.fit(X_train_1, y_train, epochs=10, batch_size=32, validation_data=(X_test_1, y_test))
```

```python
# Оценка модели
loss, mae, mse = model.evaluate(X_test_1, y_test)
print(f"Test Loss: {loss}")
print(f"Test MAE: {mae, mse}")
```

```python
model1.fit(X_train_1, y_train, epochs=25, batch_size=256, validation_data=(X_test_1, y_test))
# Оценка модели
loss, mae, mse = model1.evaluate(X_test_1, y_test)
print(f"Test Loss: {loss}")
print(f"Test metrics: {mae, mse}")
```

```python
import tensorflow as tf

# Создание модели
model1 = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(10, 10, 10, 1)),
    
    # Первый блок свёртки
    tf.keras.layers.Conv3D(filters=32, kernel_size=3, activation='elu', padding='same'),
    tf.keras.layers.Conv3D(filters=32, kernel_size=3, activation='elu', padding='same'),  # Дополнительный свёрточный слой
    tf.keras.layers.MaxPooling3D(pool_size=2),
    tf.keras.layers.Dropout(0.2),
    
    # Второй блок свёртки
    tf.keras.layers.Conv3D(filters=64, kernel_size=3, activation='elu', padding='same'),
    tf.keras.layers.Conv3D(filters=64, kernel_size=3, activation='elu', padding='same'),  # Дополнительный свёрточный слой
    tf.keras.layers.MaxPooling3D(pool_size=2),
    tf.keras.layers.Dropout(0.2),
    
    # Третий блок свёртки
    tf.keras.layers.Conv3D(filters=128, kernel_size=2, activation='elu', padding='same'),  # Уменьшено ядро свёртки до 2
    tf.keras.layers.MaxPooling3D(pool_size=2),
    tf.keras.layers.Dropout(0.2),
    
    # Переход к полносвязным слоям
    tf.keras.layers.Flatten(),
    
    # Полносвязные слои
    tf.keras.layers.Dense(units=256, activation='elu'),  # Увеличено количество нейронов
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(units=128, activation='elu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(units=64, activation='elu'),
    tf.keras.layers.Dropout(0.2),
    
    # Выходной слой
    tf.keras.layers.Dense(units=21)  # Выходной слой для регрессии, 21 число
])

# Компиляция модели
model1.compile(optimizer='nadam', loss='mean_absolute_error', metrics=['mean_absolute_error', 'mean_squared_error'])

# Вывод структуры модели
model1.summary()
```

```python
model1.fit(X_train_1, y_train, epochs=25, batch_size=256, validation_data=(X_test_1, y_test))
# Оценка модели
loss, mae, mse = model1.evaluate(X_test_1, y_test)
print(f"Test Loss: {loss}")
print(f"Test metrics: {mae, mse}")
```
